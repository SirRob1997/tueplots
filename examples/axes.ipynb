```python
from tueplots import axes
import matplotlib.pyplot as plt

# Increase the resolution of all the plots below
plt.rcParams.update({"figure.dpi": 150})
```

We can change the axes behaviour via `tueplots.axes`. 

```python
axes.lines()
```

```python
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
plt.show()
```

```python
plt.rcParams.update(axes.lines())
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.show()
```

```python
plt.rcParams.update(axes.tick_direction(x="inout", y="in"))
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.show()
```

```python
plt.rcParams.update(axes.color(base="red"))
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.], label="ABC")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.grid()
plt.show()
```

```python
plt.rcParams.update(axes.grid())
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.], label="ABC")
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.grid()
plt.show()
```

```python
plt.rcParams.update(axes.spines(bottom=False, right=False))
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.show()
```

```python
plt.rcParams.update(axes.color(face="darkslategray", base="red"))
fig, ax = plt.subplots()
ax.plot([1., 2.], [3., 4.])
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
ax.set_title("title")
plt.show()
```
